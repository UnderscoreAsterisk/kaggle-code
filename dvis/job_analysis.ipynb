```python
def domination(df):
    titles = df.job_title.value_counts()
    return titles[1]/titles[0] * 100
jobs.groupby(['category']).apply(domination).sort_values().plot.bar()
```

```python
def extract_salary(s):
    if '¬' in s:
        return 'approx'
    elif '£' in s:
        return 'pound'
    elif '$' in s:
        return 'USD$'
    else:
        return 'Not specified'
jobs.salary_offered.map(extract_salary).value_counts()
```

```python
def es(s):
    if 'annum' in s:
        return 'annum'
    elif 'day' in s:
        return 'day'
    elif 'hour' in s:
        return 'hour'
    elif 'year' in s:
        return 'year'
    else:
        return 'Other'
jobs.salary_offered.map(es).value_counts().plot.bar()
```
